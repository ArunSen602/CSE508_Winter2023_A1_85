```python
def extract_text(fname):
  fobj=open(fname,'r')
  contents=fobj.read()
  sp=BeautifulSoup(contents,'lxml') 
  #title_txt=sp.title.text
  #text_txt=sp.TEXT.text
  text_txt=''
  title_txt=''
  for txt in sp.find_all('text'):
    text_txt=txt.get_text()
  for title in sp.find_all('title'):
    title_txt=title.get_text()
  fin_txt=''.join([title_txt,text_txt])
  #print(fin_txt)
  fobj.close()
  f_wrt=open(fname,'w')
  f_wrt.write(fin_txt)
  f_wrt.close()
```

```python
import os
files_list=os.listdir('/content/temp_dir/CSE508_Winter2023_Dataset')
dir_path='/content/temp_dir/CSE508_Winter2023_Dataset'
for fl in files_list:
  fl_path=os.path.join(dir_path,fl)
  extract_text(fl_path)
```

```python
import re
for fl in files_list:
  fl_path=os.path.join(dir_path,fl)
  fobj=open(fl_path,'r+')
  text=fobj.read()
  words=re.sub('[^\w\s]',' ',text)
  fobj.seek(0)        # absolute file positioning
  fobj.truncate()     # to erase all data 
  fobj.write(words)
  fobj.close()
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words=list(stopwords.words("english"))

for fl in files_list:
  fl_path=os.path.join(dir_path,fl)
  fobj=open(fl_path,'r+')
  text=fobj.read()
  reexp='|'.join(stop_words)
  reexp="\W"+"("+reexp+")"+"\W"
  text_new=re.sub(reexp,' ',text)
  fobj.seek(0)        # absolute file positioning
  fobj.truncate()     # to erase all data 
  fobj.write(text_new)
  fobj.close()
```

```python
import nltk
nltk.download('punkt')
for fl in files_list:
  fl_path=os.path.join(dir_path,fl)
  fobj=open(fl_path,'r+')
  text=fobj.read()
  words=nltk.word_tokenize(text)
  new_text = ",".join(words)
  fobj.seek(0)        # absolute file positioning
  fobj.truncate()     # to erase all data 
  fobj.write(new_text)
  fobj.close()
```
