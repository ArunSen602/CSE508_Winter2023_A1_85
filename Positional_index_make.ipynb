```python
import os
from zipfile import ZipFile
import nltk
nltk.download('punkt')
class Positional_Index:
  def __init__(self):
    self.index_dict=dict()
    self.token_list=list()

  def make_tok_list(self):
    os.mkdir('temp_dir1')
    with ZipFile("/content/CSE508_Winter2023_Dataset_1.zip", 'r') as zObject:    # Extracting all the members of the zip 
    # into a specific location.
      zObject.extractall(path="temp_dir1")
    dir_path='/content/temp_dir1/content/temp_dir/CSE508_Winter2023_Dataset'
    files_list=os.listdir(dir_path)
    file_tok_list=list()
    for fl in files_list:
      fl_path=os.path.join(dir_path,fl)
      fobj=open(fl_path,'r')
      st=fobj.read()
      word_toks=nltk.word_tokenize(st)
      self.token_list.append(word_toks)
      fobj.close()

  def make_index(self):
    fin_token_list=set()
    #index_dict=dict()
    self.make_tok_list()
    for toks in self.token_list:
      fin_token_list=fin_token_list.union(set(toks))
    fin_token_list=list(fin_token_list)
    for tok in fin_token_list:
      self.index_dict[tok]=dict()
    for doc in range(len(self.token_list)):
      for pos in range(len(self.token_list[doc])):
        if doc not in self.index_dict[self.token_list[doc][pos]].keys():
          self.index_dict[self.token_list[doc][pos]][doc]=[]
          self.index_dict[self.token_list[doc][pos]][doc].append(pos)
        else:
          self.index_dict[self.token_list[doc][pos]][doc].append(pos)
      #index_dict[self.token_list[doc][pos]]
    return self.index_dict

  def process_query(self,query_text):
    q_toks=nltk.word_tokenize(query_text)
    pos_list_dict=dict()
    for i in range(len(q_toks)):
      for doc in self.index_dict[q_toks[i]].keys():
        print(doc)
        if i==0 and (doc not in pos_list_dict.keys()):
          pos_list=self.index_dict[q_toks[i]][doc]
          pos_list_dict[doc]=set()
          pos_list_dict[doc]=pos_list_dict[doc].union(set(pos_list))
        elif doc in pos_list_dict.keys():
          pos_list=self.index_dict[q_toks[i]][doc]
          pos_list=[(pos-i)  for pos in pos_list]
          pos_list_dict[doc]=pos_list_dict[doc].intersection(set(pos_list))
        else:
          if doc in pos_list_dict:
            del pos_list_dict[doc]
    return pos_list_dict,len(pos_list_dict.keys())
```
